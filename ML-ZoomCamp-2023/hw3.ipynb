## Mutual Info

- Calculate the mutual information score between above_average and other categorical variables in our dataset. Use the training set only.
- Round the scores to 2 decimals using round(score, 2).

```python
from sklearn.metrics import mutual_info_score

mutual_info_score(above_average_train, df_train.price)
```

```python
mutual_info_score(above_average_train, df_train.year)
```

```python
def mutual_info_churn_score(series):
    return mutual_info_score(series, above_average_train)
```

```python
mi = df_train[categorical_columns].apply(mutual_info_churn_score)
mi.sort_values(ascending=False).round(2)
```

```python
#Import math Library
import math 

for c in categorical_columns:
    categ = categorical_columns.copy()
    categ.remove(c)
    print(categ)
    train_dict = df_train[categ + numerical_columns].to_dict(orient='records')
    dv_new = DictVectorizer(sparse=False)
    X_train_new = dv_new.fit_transform(train_dict)
    print(f"X_train_new.shape: {X_train_new.shape}")
    val_dicts = df_val[categ + numerical_columns].to_dict(orient='records')
    X_val_new = dv_new.transform(val_dicts)
    model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
    y_train = above_average_train
    model.fit(X_train_new, y_train)
    y_pred_new = model.predict_proba(X_val_new)[:, 1]
    churn_decision_new = (y_pred_new >=0.5)
    y_val = above_average_val
    acc_new = (y_val == churn_decision_new).mean()
    print(f"Accuracy: {acc_new} diff: {math.fabs(accuracy-acc_new)}")
```

```python
for alpha in [0, 0.01, 0.1, 1, 10]:
    print(f"processing for alpha: {alpha}")
    model = Ridge(alpha=alpha, solver='sag', random_state=42, max_iter=200)
    model.fit(X_train, log_y_train)
    y_pred = model.predict(X_val)
    y_val = df_val['price']
    print(f"y_pred.shape: {y_pred.shape} type: {type(y_pred)}, y_val.shape: {y_val.shape} type: {type(y_val)}")
    accuracy = mean_squared_error(y_val, y_pred, squared=False)
    print(f"Alpha: {alpha}, Accuracy: {accuracy}")
```
